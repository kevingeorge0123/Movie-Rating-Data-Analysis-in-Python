```python
list1= list()  # same as list1=[] ie. empty list
mylabels=list()
for gen in movies.Genre.cat.categories:
    list1.append(movies[movies.Genre==gen].BudgetMillions)
    mylabels.append(gen)

sns.set_style('whitegrid')    
fig, ax =plt.subplots()   # creating plot on subplot
fig.set_size_inches(11.7,8.27)  #size of A4
plt.hist(list1,stacked=True,rwidth=0.9,label=mylabels)
plt.title('Movie Budget Distribution',fontsize=35,color='darkblue',fontname='Console')
plt.ylabel('Number of movies',fontsize=25,color='red')
plt.xlabel('Budget',fontsize=25,color='green')
plt.yticks(fontsize=20)
plt.xticks(fontsize=20)
plt.legend(frameon=True,shadow=True,framealpha=1,prop={'size':20})   # or plt.legend(loc='upper left',bbox_to_anchor=[1,1])
plt.show()
```
